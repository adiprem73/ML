# Series

### Panda series = like a table column - 1 dimensional array holding any type of data

```python
import pandas as pd
```

```python
my_series = [5,9,12,27]
my_var = pd.Series(my_series)
my_var
```

```python
# print specific value
my_var[2]
```

```python
# labes using the index arguement
my_index = ["A","B","C","D"]
my_var2 = pd.Series(my_series, my_index)
my_var2
```

```python
my_var2["B"]
```

```python
# key valye dictionary
cars ={"Tesla":12, "Mercedes":40, "BMW":3}
my_var3 = pd.Series(cars)
my_var3
```
