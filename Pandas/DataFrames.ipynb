# Count Data

```python
my_df2
```

```python
dog= pd.read_csv('dog_data.csv')
dog
```

```python
# count disting values - descending
dog['Color'].value_counts()
```

```python
# count distinct values- ascending
dog['Color'].value_counts(ascending=True)
```

```python
# dog names - without NaN
dog['DogName'].value_counts()
```

```python
# dog names - with NaN
dog['DogName'].value_counts(dropna=False)
```

```python
# get relative frequency - Percentage
dog['Color'].value_counts(normalize=True)
```

```python
# get specified item count
dog['Color'].value_counts()['WHITE']
```

```python
# count unique values - size
dog.groupby('Color').size()
```

```python
dog.groupby('Color').count()
```

```python
# get a count of all cols across all cols
dog.apply(pd.value_counts)
```

## Add columns in Data Frames


```python
my_df2
```

```python
# add column from list
gender = ["male","female", "male", "female"]
my_df2["Gender"]= gender
my_df2
```

```python
# add default values
my_df2["Alive/Dead"] = [True]*len(my_df2)
my_df2
```

```python
# add NaN values
my_df2["Show Dog"]= [np.nan]*len(my_df2)
my_df2
```

```python
# add column with .insert() = allows position
my_df2.insert(1, "Adopted", [True]*len(my_df2), True); # position, name of the colum, wha tot insert, default
my_df2
```

```python
# add column with .assign() - creates a new df - cretaes a new datafram so that the old one is secured and unchaned

my_df3 = my_df2.assign(Horse = [False]*len(my_df2))
my_df3
```
