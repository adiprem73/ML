# Group By

```python
import pandas as pd
import numpy as np
# create data
stuff = {
    'Corporation':['Apple','Google','Meta','Apple','Google','Meta'],
    'Employees':['John','April','Wes','Beth','Justin','Steph'],
    'Salary':[200,220,190,130,120,150]
}

# create datafram
my_df = pd.DataFrame(stuff)
my_df
```

```python
# group by corporation - to get object locaion in memory
company = my_df.groupby('Corporation')
company
```

```python
# sum
company.sum()
```

```python
# mean
company['Salary'].mean()
```

```python
company.max()
```

```python
company.std()
```

```python
company['Salary'].std()
```

```python
# describe
company.describe()
```
