# Apply function - All Powerful

### allows uou tro applyu any kind of function to any data in a dataframe. anything i want to do in python,, write a function and then use it in our data

```python
import pandas as pd
import numpy as np
# create data
stuff = {
    'Corporation':['Apple','Google','Meta','Apple','Google','Meta'],
    'Employees':['John','April','Wes','Beth','Justin','Steph'],
    'Salary':[200,220,190,130,120,150]
}

# create datafram
my_df = pd.DataFrame(stuff)
my_df
```

```python
# create function
def times1000(x):
    return format(x*1000, ",d")
```

```python
# use apply function
pd.DataFrame(my_df["Salary"].apply(times1000))
```

```python
# add elder alst name
def namer(x):
    if x=="John":
        return "John Elder"
    else:
        return "Not important"
```

```python
my_df["Employees"].apply(namer)
```

```python
# use lambda functions: write the functions in place in line and no need to write them seperated
pd.DataFrame(my_df["Salary"].apply(lambda x : x*1000))
```
