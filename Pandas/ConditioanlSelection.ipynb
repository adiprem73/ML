# Conditional Selection

```python
import numpy as np
import pandas as pd
# import csv file
my_df = pd.read_csv('animal.csv')
my_df2 = pd.read_csv('dog_data.csv')
my_df2
```

```python
# conditioan selection
# > < = <= >= etc
# return booleans
my_df2== "BROWN"
```

```python
# returns datafram with data
my_df2[my_df2== "BROWN"]
```

```python
# run them on a column
my_df2[my_df2 == "BROWN"]["Color"]
```

```python
my_df2[my_df2 == "BROWN"][["Color", "Breed"]]
```

## Multiple conditional selections

```python
# multiple conditinal AND
my_df2[(my_df2["Color"]=="BROWN") & (my_df2["Breed"]=="MIXED")]
```

```python
# get length
len(my_df2[(my_df2["Color"]=="BROWN") & (my_df2["Breed"]=="MIXED")])
```

```python
# multiple conditoinal OR
my_df2[(my_df2["Color"]=="BROWN") | (my_df2["Breed"]=="MIXED")]
```

```python
my_df2[(my_df2["Color"]=="BROWN") & (my_df2["Breed"]=="MIXED")]["DogName"]
```
