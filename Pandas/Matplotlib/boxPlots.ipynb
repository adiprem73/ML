# box plots

```python
import pandas as pd
import numpy as np
from numpy.random import randn
%matplotlib inline

# Create a dataframe
my_df = pd.DataFrame(randn(500, 4), columns = ["Mon","Tue","Wed", "Thur"])
my_df
```

```python
my_df.plot(kind="box")
```

```python
my_df.plot(kind="box", colormap="jet")
```
