# Line plot using matplotlib


```python
import pandas as pd
import numpy as np
from numpy.random import randn
%matplotlib inline

# Create a dataframe
my_df = pd.DataFrame(randn(25, 4), columns = ["Mon","Tue","Wed", "Thur"])
my_df
```

```python
my_df.abs().plot(kind = "line")
```

```python
# line width - lw
my_df.abs().plot(kind = "line", lw=4)
```

```python
my_df["Mon"].abs().plot(kind = "line")
```

```python
my_df[["Mon", "Tue"]].abs().plot(kind = "line")
```

```python
my_df[["Mon", "Tue"]].abs().plot(kind = "line", figsize=(15,5))
```

```python
# second way 
my_df.plot.line()
```
