# Scatter plot

```python
import pandas as pd
import numpy as np
from numpy.random import randn
%matplotlib inline

# Create a dataframe
my_df = pd.DataFrame(randn(500, 4), columns = ["Mon","Tue","Wed", "Thur"])
my_df
```

```python
my_df.plot(kind="scatter", x="Mon", y= "Tue", c="Wed")
```

```python
my_df.plot(kind="scatter", x="Mon", y= "Tue", c="Wed", cmap='magma')
```

```python
# by size
my_df.plot(kind="scatter", x="Mon", y="Tue",s=my_df["Wed"]*100)
```

```python
my_df.plot(kind="scatter", x="Mon", y="Tue",s=my_df["Wed"]*100, alpha=0.4)
```
