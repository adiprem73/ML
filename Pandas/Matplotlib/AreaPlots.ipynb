# Area Plots

```python
import pandas as pd
import numpy as np
from numpy.random import randn
%matplotlib inline

# Create a dataframe
my_df = pd.DataFrame(randn(25, 4), columns = ["Mon","Tue","Wed", "Thur"])
my_df
```

```python
# area plot
my_df.abs().plot(kind = "area")
```

```python
# shading
my_df.abs().plot(kind = "area", alpha=0.4)
```

```python
# title
my_df.abs().plot(kind = "area", alpha=0.4, title="my awesom area plot")
```

```python
# legend
my_df.abs().plot(kind = "area", alpha=0.4, title="pre,", legend=False)
```

```python
# plot one column only
my_df["Mon"].abs().plot(kind = "area", alpha=0.4, title="pre,", legend=False)
```

```python
# plot 2 cols
my_df[["Mon", "Thur"]].abs().plot(kind = "area", alpha=0.4, title="pre,", legend=False)
```

```python
# get info on function
my_df[["Mon", "Thur"]].plot(kind = "area", stacked=False, alpha=0.4, title="pre,", legend=False)
```
