# hexz Bin Plots

```python
import pandas as pd
import numpy as np
from numpy.random import randn
%matplotlib inline

# Create a dataframe
my_df = pd.DataFrame(randn(500, 4), columns = ["Mon","Tue","Wed", "Thur"])
my_df
```

```python
my_df.plot(kind="hexbin", x="Mon", y="Tue")
```

```python
my_df.plot(kind="hexbin", x="Mon", y="Tue", gridsize = 25)
```

```python
my_df.plot(kind="hexbin", x="Mon", y="Tue", C= "Wed", gridsize = 25 , colormap="magma")
```

# Densite Plots or KDE plots


```python
my_df.plot(kind="kde")
```

```python
# other way
my_df.plot.kde()
```

```python
# density plot
my_df.plot.density()
```
