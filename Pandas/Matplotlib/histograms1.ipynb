# Histographs wiht Matplotlib

```python
import pandas as pd
import numpy as np
from numpy.random import randn
%matplotlib inline

# Create a dataframe
my_df = pd.DataFrame(randn(100, 4), columns = ["Mon","Tue","Wed", "Thur"])
my_df
```

```python
# Histogram
my_df["Wed"].hist()
```

```python
# histogram with smaller bins ,50, 20
my_df["Wed"].hist(bins= 100)
```

```python
# another way withough grid bars
my_df["Wed"].plot(kind = "hist")
```

```python
my_df[["Wed", "Tue"]].hist(bins= 100, grid = False, legend = True)
```
