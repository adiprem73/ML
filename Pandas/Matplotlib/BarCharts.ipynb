# Bar Charts with Matplotlib


```python
import pandas as pd
import numpy as np
from numpy.random import randn
%matplotlib inline

# Create a dataframe
my_df = pd.DataFrame(randn(25, 4), columns = ["Mon","Tue","Wed", "Thur"])
my_df
```

```python
# bar plot
my_df.abs().plot(kind="bar")
```

```python
# stacked bar plot
my_df.abs().plot(kind="bar", stacked=True, alpha=0.3)
```

```python
# second way
my_df.abs().plot.bar(stacked=True)
```
