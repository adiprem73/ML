# Incomplete data


```python
import pandas as pd
import numpy as np
# create dummy data
stuff = {'A':[1,2,3], 'B':[4,np.nan,6], 'C':[7,8,9], 'D':[10,11,12]}
my_df = pd.DataFrame(stuff)
my_df
```

```python
# drop rows with null data
my_df.dropna()
```

```python
# drop column with null data
my_df.dropna(axis=1)
# use inplace = true to make it permanent
```

```python
# more thatn one? Set Threshold
my_df.dropna(thresh=2, axis=1)
```

```python
# replace thing with fillna()
my_df.fillna(value="Bob")
```

```python
# use math functions like filling the missing data with the average of the column
my_df.fillna(value = my_df['B'].mean())
```

```python
# use min or max or sum
my_df.fillna(value = my_df['B'].min())
```
