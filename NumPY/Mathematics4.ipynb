## Linear Algebra

```python
a = np.ones((2,3))
b= np.full((3,2),2)
np.matmul(a,b)

```

```python
# finding determinant
c = np.identity(3)
print(np.linalg.det(c))
```

## Reorganizing arrays

```python
before = np.array([[1,2,3,4],[4,5,6,7]])
print(before)

after= before.reshape((4,2))
print(after)
```

```python
# vertically stacking vecotr
v1 = np.array([1,2,3,4])
v2 = np.array([5,6,7,8])

np.vstack([v1,v2,v1,v2])
```
