# Initializing Different Types of Arays

```python
# all 0s matrix
import numpy as np
a= np.zeros((2,3,3))
```

```python
np.ones((1,2))
```

```python
np.full((2,2), 99)
```

```python
# any other number (full_like)
np.full_like(a, 4)
```

```python
# random decimal numbers
np.random.rand(4,2,3)
```

```python
# random integer values
np.random.randint(7, size=(3,3))
```

```python
np.identity(3)
```

```python
arr = np.array([[1,2,3]])
r1 = np.repeat(arr, 3, axis=0)
print(r1)
```

```python
a=np.ones((5,5))
# print(a)

z= np.zeros((3,3))
z[1,1]=9

a[1:4,1:4]=z
print(a)
```

```python
x= np.array([1,2,3])
y= x.copy()
y[1]=100
print(x)
print(y)
```
